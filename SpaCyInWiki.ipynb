```
wp=wk.WikipediaPage('Rabindranath Tagore');
WikiContent=wp.content.capitalize();
```

```

!python -m spacy download en_core_web_lg
```

```
import spacy
nlp = spacy.load("en_core_web_lg")
doc = nlp(WikiContent)
```

```
# how many words are  there?
print(len(doc))
#how many char are  there?
print(len(WikiContent))
# how many enteties are there
print(len(doc.ents))
#how many sentenses are there
print (len(list (doc.sents)))
#how many times each word  is present in the doc
from collections import Counter
words = [token.text for token in doc if (not token.is_stop and not  token.is_punct and not  token.is_space and not token.text=='=')]
word_freq= Counter(words)
print (word_freq)


#print (len(list (doc.text=="Rabindranath")))
```

```
common_words = word_freq.most_common(5)
common_words
```

```
# how many time Rabindranath is there? Because it is being used as dictionary
word_freq['rabindranath']
```

```
from spacy  import displacy 
#print(doc.ents.count("Rabindranath tagore") )
print(len(doc.ents))

displacy.render(doc, style='ent', jupyter=True,)
```

#summery 

```

# Sentence Tokens
sentence_list = [ sentence for sentence in doc.sents ]
sentence_list


```

```
# Sentence Score via comparrng each word with sentence
sentence_scores = {}  
for sent in sentence_list:  
        for word in sent:
            if word.text.lower() in word_freq.keys():
                if len(sent.text.split(' ')) < 30:
                    if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_freq[word.text.lower()]
                    else:
                        sentence_scores[sent] += word_freq[word.text.lower()]
sentence_scores
```

```
from heapq import nlargest
summarized_sentences = nlargest(7, sentence_scores, key=sentence_scores.get)
summarized_sentences
```

```
for w in summarized_sentences:
    print(w.text)
```

```
# List Comprehension of Sentences Converted From Spacy.span to strings
final_sentences = [ w.text for w in summarized_sentences ]
final_sentences
```

```
summary = ' '.join(final_sentences)
summary
```
